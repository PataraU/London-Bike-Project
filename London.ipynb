```python
#rename the columns, appropriately
london = london.rename(columns = {
    'cnt' : 'count',
    't1' : 'real_temp',
    't2' : 'feels_like_temp',
    'hum' : 'humidity_percent',
    'weather_code' : 'weather_category',
    'is_holiday' : 'holiday',
    'is_weekend' : 'weekend'
})
```

```python
london
```

```python
#find null values
london.isnull().sum()
```

```python
#change values within 'season' column from number into string
def  season(value):
    if value == 1.0:
        return 'spring'
    elif value == 2.0:
        return 'fall'
    elif value == 3.0:
        return 'winter'
    return value  # Return existing string values as is

# Apply the function to the 'season' column
london['season'] = london['season'].apply(season)

print(london)
```

```python
#change values within 'season' column from number into string
weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
}

# changing the weather column data type to string
london.weather_category = london.weather_category.astype('str')

# mapping the values to the actual written weathers
london.weather_category = london.weather_category.map(weather_dict)

print(london)
```
